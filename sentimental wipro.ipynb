## Scrapping news headlines and paragraphs

```python
ur=[]
for i in range(1,9):
    search_url='https://in.investing.com/equities/wipro-ltd-news/'+str(i)
    ur.append(search_url)
```

```python
ur
```

```python
heading = []
para=[]
date=[]

for i in ur:
    req = Request(url=i,headers={'user-agent': 'my-app/0.0.1'})
    page = urlopen(req)    
    # Read the contents of the file into 'html'
    soup = BeautifulSoup(page)
    content = soup.find('ul', {"class": "common-articles-list"})
    for i in content.findAll(['h3']):
        heading.append(i.text)
    for k in content.findAll(['p']):
        para.append(k.text)
    for j in content.findAll(['time']):
        date.append(j.text)
```

```python
articles=[]
for (f, b) in zip(heading, para):
    s=str(f)+str(b)
    articles.append(s)
    
```

## Retaining extreme (max and min) compound scores for same Day news headlines

```python
df['date'] = pd.to_datetime(pd.to_datetime(df['Date']).dt.date)
```

```python
grouped_dates = df.groupby(['Date'])

keys_dates = list(grouped_dates.groups.keys())

```

```python
a=[]
for key in grouped_dates.groups.keys():
    data = grouped_dates.get_group(key)
    a.append(data['compound'].max())
```

```python
b=[]
for key in grouped_dates.groups.keys():
    data = grouped_dates.get_group(key)
    b.append(data['compound'].min())
```

```python
max_cs=[]
min_cs=[]
```

```python
for i in a:
    if i > 0:
        max_cs.append(i)
    elif i <= 0:
        max_cs.append(0)
```

```python
for i in b:
    if i <= 0:
        min_cs.append(i)
    elif i > 0:
        min_cs.append(0)
```

```python
extreme_scores_dict = {'Date':keys_dates,'max_scores':max_cs,'min_scores':min_cs}
extreme_scores_df = pd.DataFrame(extreme_scores_dict)
```

```python
extreme_scores_df
```

## Retaining extreme (max and min) compound scores for same Day news headlines

```python
final_scores = []
for i in range(len(extreme_scores_df)):
    final_scores.append(extreme_scores_df['max_scores'].values[i] + extreme_scores_df['min_scores'].values[i])

extreme_scores_df['final_scores'] = final_scores

extreme_scores_df.head()
```

```python
sum(final_scores)/len(final_scores)
```

```python
extreme_scores_df['Date'] = pd.to_datetime(pd.to_datetime(extreme_scores_df['Date']).dt.date)
```

## Generating Trade calls using vader scores

```python
vader_Buy=[]
vader_Sell=[]
for i in range(len(extreme_scores_df)):
    if extreme_scores_df['final_scores'].values[i] > 0.30:
        print("Trade Call for {row} is Buy.".format(row=extreme_scores_df['Date'].iloc[i].date()))
        vader_Buy.append(extreme_scores_df['Date'].iloc[i].date())
    elif extreme_scores_df['final_scores'].values[i] < 0.30:
        print("Trade Call for {row} is Sell.".format(row=extreme_scores_df['Date'].iloc[i].date()))
        vader_Sell.append(extreme_scores_df['Date'].iloc[i].date())
```

```python
vader_buy = []
for i in range(len(wipro)):
    if wipro.index[i]in vader_Buy:
        vader_buy.append(i)
        
vader_sell = []
for i in range(len(wipro)):
    if wipro.index[i] in vader_Sell:
        vader_sell.append(i)
        
plt.figure(figsize=(20, 10),dpi=80)
plt.plot(wipro.index, wipro['Close'],'-^', markevery=vader_buy, ms=15, color='green')
plt.plot(wipro.index, wipro['Close'],'-v', markevery=vader_sell, ms=15, color='red')
plt.plot(wipro.index, wipro['Close'])
plt.xlabel('Date',fontsize=14)
plt.ylabel('Price in Dollars', fontsize = 14)
plt.xticks(rotation='60',fontsize=12)
plt.yticks(fontsize=12)
plt.title('Trade Calls - VADER', fontsize = 16)
plt.legend(['Buy','Sell','Close'])
plt.grid()
plt.show() 
```

## Using Simple Moving Crossover Strategy to generate trade calls 

```python
wipro['2_SMA'] = wipro['Close'].rolling(window=2).mean()
wipro['5_SMA'] = wipro['Close'].rolling(window=5).mean()

wipro = wipro[wipro['5_SMA'].notna()]

# SMA trade calls
Trade_Buy=[]
Trade_Sell=[]
for i in range(len(wipro)-1):
    if ((wipro['2_SMA'].values[i] < wipro['5_SMA'].values[i]) & (wipro['2_SMA'].values[i+1] > wipro['5_SMA'].values[i+1])):
        print("Trade Call for {row} is Buy.".format(row=wipro.index[i]))
        Trade_Buy.append(wipro.index[i])
    elif ((wipro['2_SMA'].values[i] > wipro['5_SMA'].values[i]) & (wipro['2_SMA'].values[i+1] < wipro['5_SMA'].values[i+1])):
        print("Trade Call for {row} is Sell.".format(row=wipro.index[i]))
        Trade_Sell.append(wipro.index[i])
```

```python
Trade_buy = []
for i in range(len(wipro)):
    if wipro.index[i]in Trade_Buy:
        Trade_buy.append(i)
        
Trade_sell = []
for i in range(len(wipro)):
    if wipro.index[i] in Trade_Sell:
        Trade_sell.append(i)
        
plt.figure(figsize=(20, 10),dpi=80)
plt.plot(wipro.index, wipro['2_SMA'],'-^', markevery=Trade_buy, ms=15, color='green')
plt.plot(wipro.index, wipro['5_SMA'],'-v', markevery=Trade_sell, ms=15, color='red')



plt.plot(wipro.index, wipro['Close'])
plt.xlabel('Date',fontsize=14)
plt.ylabel('Price in Dollars', fontsize = 14)
plt.xticks(rotation='60',fontsize=12)
plt.yticks(fontsize=12)
plt.title('Trade Calls - Moving Average', fontsize = 16)
plt.legend(['2_SMA','5_SMA','Close'])
plt.grid()
plt.show() 
```
